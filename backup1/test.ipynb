```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import random
import numpy as np

driver = webdriver.Firefox()
URL = 'https://uk.webuy.com/boxsearch?categoryIds=1141'
driver.get(URL)
```

```python
containers = driver.find_elements_by_xpath("/html/body/div[1]/div/div/div[6]/div[1]/div[2]/div/div/div/div[3]/div")
```

```python
for i in containers:
    try:
        id = (i.get_attribute("data-insights-object-id"))
        picture = i.find_element_by_tag_name("img")
        picture_link = picture.get_attribute("src")
        if id == None:
            break
        else:
            print(id, picture_link)
    except NoSuchElementException:
        break
```

```python
final_prc = []
names_txt = []

prices = driver.find_elements_by_class_name("prodPrice")
names_ele = driver.find_elements_by_xpath('//span[@class="ais-highlight"]')

for count, i in enumerate(prices):
    prce = i.find_elements_by_class_name("priceTxt")
    if prce == []:
        prices.remove(i)


for count, i in enumerate(prices):
    list1 = i.find_elements_by_class_name("priceTxt")
    if len(list1) != 3:
        prices.remove(prices[count])
        names_ele.remove(names_ele[count])

for i in prices:
    list1 = i.find_elements_by_class_name("priceTxt")
    for i in list1:
        price_text = i.text.split(" ")
        final_prc.append(price_text)

selling = final_prc[::3]
buy_cash = final_prc[1::3]
buy_voucher = final_prc[2::3]

for i in names_ele:
    
    name = (i.text.split(" (No DLC)")[0])
    names_txt.append(name)

id_cont = driver.find_elements_by_xpath("/html/body/div[1]/div/div/div[6]/div[1]/div[2]/div/div/div/div[3]/div")

ids = []
for i in id_cont:
    id = i.get_attribute("data-insights-object-id")
    ids.append(id)
```
